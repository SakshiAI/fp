```python
import pandas as pd
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = datasets.load_iris()
X = iris.data  # Features: sepal length, sepal width, petal length, petal width
```

```python
# 2. Optional: Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# 3. Apply KMeans Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_scaled)
```

```python

# 4. Get predicted cluster labels
cluster_labels = kmeans.labels_
```

```python
# 5. Visualize the clusters
df = pd.DataFrame(X, columns=iris.feature_names)
df['Cluster'] = cluster_labels
```

```python
# Use only two features for plotting (for 2D visualization)
sns.scatterplot(data=df, x=iris.feature_names[0], y=iris.feature_names[2], hue='Cluster', palette='Set1')
plt.title("KMeans Clustering on Iris Dataset")
plt.show()
```
